Problem Statement :-
	A cloth manufacturing company is interested to know about the different attributes contributing to high sales. Build a decision tree & random forest model with Sales as target variable (first convert it into categorical variable).

Maximize Sales Revenue: The primary objective is to increase sales revenue for the cloth manufacturing company. This involves identifying key attributes that contribute to high sales and leveraging them effectively.



Minimize Production Costs: While maximizing sales revenue is crucial, it's also important to keep production costs in check to ensure profitability. Identifying attributes that can help optimize production processes and reduce costs is essential.

```python
"""
Constraints :
    'Sales',
    'CompPrice', 
    'Income', 
    'Advertising', 
    'Population', 
    'Price',
    'ShelveLoc', 
    'Age', 
    'Education', 
    'Urban', 
    'US'
"""
```

```python
import pandas as pd
import numpy as np  
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('D:\\1-Data Science\\6 - Machine Learning\\Decision Tree\\Data\\Company_data.csv')
```

```python
df
```

```python
bin_target = [float('-inf'), 8, float('inf')]
bin_lable = ['low','high']

df['Sales'] = pd.cut(df['Sales'], bins=bin_target, labels=bin_lable, right = False)
```

```python
df.columns
```

```python
df
```

```python
le = LabelEncoder()
```

```python
l = ['ShelveLoc','Urban', 'US']

for i in l:
    df[i] = le.fit_transform(df[i])
```

```python
df
```

```python
input = df.drop('Sales', axis = 'columns')
target = df['Sales']
```

```python
X_train, X_test , y_train, y_test = train_test_split(input, target , test_size=0.2)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
model.score(X_test, y_test)
```
